```python
import sagemaker
from sagemaker.xgboost.estimator import XGBoost
from sagemaker import get_execution_role
from project_paths import PATHS, s3_uri

# Create custom session
# sess = sagemaker.Session(
#     default_bucket="sagemaker-us-east-oe-bi-temp-data",
#     default_bucket_prefix=PATHS.temp_prefix
# )

role = get_execution_role()
# role = 'arn:aws:iam::719805443631:role/bi_sagemaker'
print(f"Using role: {role}")
print(f"Training data: {s3_uri(PATHS.test_training_data_key)}")
print(f"Model prefix: {s3_uri(PATHS.model_prefix)}")

estimator = XGBoost(
    entry_point="train.py",
    source_dir="src",
    role=role,
    instance_count=1,
    instance_type="ml.m5.large",
    framework_version="1.7-1",
    output_path=s3_uri(PATHS.model_prefix),
    # sagemaker_session=sess,
)

print(f"estimator.sagemaker_session.default_bucket(): {estimator.sagemaker_session.default_bucket()}")
print(f"estimator.sagemaker_session.default_bucket_prefix: {estimator.sagemaker_session.default_bucket_prefix}")


estimator.fit({
    "train": s3_uri(PATHS.test_training_data_key)
})

print(f"estimator.model_data: {estimator.model_data}")
```
