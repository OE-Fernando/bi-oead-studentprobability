```python
import boto3
import json
import numpy as np

endpoint_name = "oead-prob-ep2026-05-06-15-17-47"

runtime = boto3.client("sagemaker-runtime")
```

```python
payload = json.dumps({
    "dow": "7",
    "studLevel": "1",
    "stuH": "4-8",
    "country_iso": "BR",
    "enrollment": "sMax",
    "native_language": "pt-BR",
    "class_type": "group",
    "ageGroup": 2,
    "deltaDays": 0,
    "deltaHours": -1,
    "hourOfDay": 7,
    "minuteOfHour": 0,
    "isWeekend": 1,
    "isHoliday": 1,
    "isHolidayPre": 0,
    "isHolidayPost": 0
})
payload
```

```python
response = runtime.invoke_endpoint(
    EndpointName=endpoint_name,
    ContentType="application/json",
    Body=payload
)
```

```python
result = response["Body"].read().decode("utf-8")
print("Raw result:", result)
```
